```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("sales_data.csv")  
```

```python
print("First 5 Rows of the Dataset:")
display(df.head())
```

```python
print("Missing Values in Dataset:")
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True) 
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
df['Date'] = pd.to_datetime(df['Date'])  
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Sales_Amount", data=df, marker="o", linewidth=2)
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="Category", y="Sales_Amount", data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales Amount")
plt.show()
```

```python
top_products = df.groupby("Product_Name")["Sales_Amount"].sum().sort_values(ascending=False).head(5)
print("Top 5 Performing Products:")
print(top_products)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="Region", y="Sales_Amount", data=df)
plt.title("Sales Distribution Across Regions")
plt.xlabel("Region")
plt.ylabel("Sales Amount")
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales_Amount'].sum()

# Create a basic pie chart
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%')

# Set title
plt.title("Sales Distribution by Region")

# Show the plot
plt.show()
```
